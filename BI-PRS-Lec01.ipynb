### Pojmy z BI-PST

Připomeňme některé důležité pojmy z BI-PST. Cílem je uvědomit si, že naše *statistické* analýzy se budou číselně lišit i když se budeme zabývat toutéž veličinou. A to proto, že pracujeme s realizací *výběru z náhodné veličiny*, jež je ryze teoretickým pojmem z pravděpodobnosti (měřitenou funkcí). Budeme uvažovat, že náhodná veličina $X$ má distribuční funkci $F$.

> **Náhodný výběr**
>
> $n$-tice stejně rozdělených nezávislých veličin $X_1,\ldots, X_n$ s distribuční funkcí $F$ se nazývá **náhodný výběr** z rozdělení $F$.

Tento pojem je stále ještě dost abstraktní, my potřebujeme pracovat s konkrétními čísly - realizací:

> **Realizace náhodného výběru** (též **data**) je $n$-tice konkrétních pozorovaných čísel $x_1,\ldots,x_n$.


Několikrát si vygenerujme realizaci náhodného výběru o rozsahu $n=10$ z rozdělení $F = \mathit{Pois}(15)$ a diskutujme o očekávaných vlastnostech v souvislosti s pojmy *rozsah výběru*, *zákon velkých čísel* aj.

```r
sample1 <- rpois(10, lambda=15); print(sample1)
sample2 <- rpois(10, lambda=15); print(sample2)
sample3 <- rpois(10, lambda=15); print(sample3)
```

## Skalární veličiny (univariate analysis)

Uvažujme prostý náhodný výběr jednorozměrných iid veličin. Zrekapitulujme významné statistické pojmy zavedené v [BI-PST](https://courses.fit.cvut.cz/BI-PST/).

### Míry polohy

- [**Výběrový průměr**](https://en.wikipedia.org/wiki/Sample_mean_and_covariance)
$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$
- [**Modus**](https://en.wikipedia.org/wiki/Mode_(statistics)) $\hat{X}$ je definován jako lokální maximum na hustotě pravděpodobnosti nebo pravděpodobnostní funkci. V případě dat je modem nejčetnější hodnota. 
- [**Kvantily**](https://en.wikipedia.org/wiki/Quantile) jsou populární v [robustní statistice](https://en.wikipedia.org/wiki/Robust_statistics) pro svou relativní odolnost vůči odchylkám od normality, [outlierům](https://en.wikipedia.org/wiki/Outlier) apod.
    > Nechť $X$ je náhodná veličina s distribuční funkcí $F(x)$. Dále nechť $\alpha \in (0, 1)$. **$\alpha$-kvantilem** veličiny $X$ je bod $q_{\alpha} = \inf\{x: F(x) \geq \alpha\}$.
    - **Kvartily** - dolní kvartil $q_{0.25}$, medián $q_{0.5}$ a horní kvartil $q_{0.75}$. V případě náhodného výběru provedeme jeho *uspořádání* podle velikosti $X_{(1)} \leq X_{(2)}\leq \ldots \leq X_{(n)}$. Medián rozděluje výběr na dvě stejně velké poloviny, tj.
    $$
    q_{0.5} = 
    \begin{cases}
        X_{(n+1)/2} \quad &\text{pro $n$ liché},\\ 
        \frac{1}{2} \left( X_{\left(\frac{n}{2}\right)} + X_{\left(\frac{n+1}{2}\right)} \right) \quad &\text{pro $n$ sudé}.
    \end{cases}
    $$
    Analogicky dolní a horní kvartil oddělují dolní a horní čtvrtinu dat.
    - **Decily, percentily** a drobnější kvartily dělí uspořádaný výběr do deseti, sta, event. více podsouborů.
    

### Míry rozptýlení
- [**Výběrový rozptyl**](https://en.wikipedia.org/wiki/Variance#Sample_variance)
$$
S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})^2
$$
- **Výběrová směrodatná odchylka**
$$
S = \sqrt{S^2}
$$
- **Rozpětí** - označíme-li $X_{\text{min}}$ a $X_{\text{max}}$ minimum a maximum z náhodného výběru,
$$
R = X_{\text{max}} - X_{\text{min}}
$$
- [**Mezikvartilové rozpětí (IQR)**](https://en.wikipedia.org/wiki/Interquartile_range) - [robustní](https://en.wikipedia.org/wiki/Robust_statistics) alternativa rozptylu.
$$
\mathrm{IQR} = q_{0.75} - q_{0.25}.
$$

### Míry šikmosti a špičatosti

[Šikmost](https://en.wikipedia.org/wiki/Skewness) a [špičatost](https://en.wikipedia.org/wiki/Kurtosis) jsou definovány jako podíly vyšších momentů (viz BI-PST). Níže uvedené výběrové statistiky jsou ovšem vychýlené, proto se často používají různé modifikace. Pro praktické účely je podstatné porovnání s normálním rozdělením, jehož šikmost i špičatost nabývají nulových hodnot.

- **Výběrová šikmost**
$$
g_{1} = \frac{\frac{1}{n}\sum_{i=1}^{n} (X_i - \bar{X})^3}{\left[\frac{1}{n-1}\sum_{i=1}^{n} (X_i - \bar{X})^2\right]^{3/2}}
$$

- **Výběrová špičatost**
$$
g_{2} = n \frac{\sum_{i=1}^{n} (X_i - \bar{X})^4}{\left[\sum_{i=1}^{n} (X_i - \bar{X})^2\right]^{2}} \color{red}{- 3}
$$

### Příklad

> [**Beta distribuce**](https://en.wikipedia.org/wiki/Beta_distribution)
>
> Říkáme, že spojitá náhodná veličina $X$ s hodnotami na intervalu [0, 1] má rozdělení beta s parametry $a,b>0$, pokud její hustota pravděpodobnosti je dána předpisem
>
>$$
f(x; a,b) = 
\begin{cases}
\frac{\Gamma(a)\Gamma(b)}{\Gamma(a+b)} x^{a-1} (1-x)^{b-1} \qquad &\text{pro $x\in[0,1]$}, \\
0 &\text{jinde}.
\end{cases}
$$

U beta distribuce je snadné měnit šikmost a špičatost pomocí parametrů $a$ a $b$. Zkuste experimentovat v kódu, všímejte si, jak se mění tvar hustoty a výběrové statistiky:

```r
library(ggplot2)
library(cowplot)
library(vtable)

a <- 2
b <- 2
x <- rbeta(100000, a, b)
df <- data.frame(x)
summ <- c('mean(x)', 'var(x)', 'min(x)', 'pctile(x)[25]', 'median(x)', 'pctile(x)[75]', 
          'max(x)', 'IQR(x)', 'skewness(x)', 'kurtosis(x)')
print.table(sumtable(df, out="return", summ=summ, ))

ggplot(data = data.frame(x = c(-.5, 1.5)), aes(x)) +
    stat_function(fun=dbeta, args=list(a, b), n=101, colour="red")
```

### Tvar distribuce

Tvar distribuce poměrně dobře přiblíží kvantily s dostatečnou jemností (decily a zejména percentily), nejlépe jej ale definují
- [hustota pravděpodobnosti](https://en.wikipedia.org/wiki/Probability_density_function) - pdf - v případě spojité veličiny
- [pravděpodobnostní funkce](https://en.wikipedia.org/wiki/Probability_mass_function) - pmf - v případě diskrétní veličiny,
- [distribuční funkce](https://en.wikipedia.org/wiki/Cumulative_distribution_function) - cdf

V teorii pravděpodobnosti v BI-PST jsme se setkali s příslušnými definicemi pro náhodnou veličinu. V BI-PRS se nicméně zabýváme *realizacemi náhodného výběru*, tedy **daty**. Namísto pdf, pmf a cdf tedy budeme hledat jejich odhady, též empirické pdf, pmf a cdf.

#### Empirická distribuční funkce
> Buď $X_{(1)}, \ldots, X_{(n)}$ uspořádaný náhodný výběr z rozdělení s distribuční funkcí $F(x)$. Empirická distribuční funkce je funkce ve tvaru
>
>$$
\hat{F}_n(x) = 
\begin{cases}
0 &\text{pro $x<X_{(1)}$},\\
i/n &\text{pro $X_{(i)} \leq x \leq X_{(i+1)}$}\qquad (i=1,\ldots,n-1),\\
1 &\text{pro $x\geq X_{(n)}$}.
\end{cases}
$$
>
V BI-PST není prostor na důkazy, spokojme se aspoň s tím, že podle zákona velkých čísel $\hat{F}_{n}(x) \xrightarrow{n\to \infty} F(x)$ pro každé $x\in\mathbb{R}$.

```r
x <- rnorm(15)
plot.ecdf(x)
```

#### Odhad hustoty/pravděpodobnostní funkce

Tvar hustoty či pravděpodobnostní funkce odhadujeme zejména pomocí 
- [histogramu](https://en.wikipedia.org/wiki/Histogram) - tato problematika je dobře známá, je zejména potřeba vybrat vhodný počet přihrádek (metodou pokus-omyl, Sturgesovým pravidlem apod.),
- [jádrového odhadu hustoty](https://en.wikipedia.org/wiki/Kernel_density_estimation) - pomocí vhodné funkce (jádra) konstruujeme hladký odhad hustoty.

> **Jádrový odhad hustoty (Kernel density estimate, KDE)**
>
>Buď $X_1, \ldots, X_n$ náhodný výběr z rozdělení s hustotou $f(x)$. Jádrovým odhadem hustoty nazýváme funkci
>
>$$
\hat{f}_n(x) = \frac{1}{nh} \sum_{i=1}^{n} K\left(\frac{x - x_i}{h}\right)
$$
>
>kde $K()$ je symetrická hustota pravděpodobnosti nazývaná [*jádrem*](https://en.wikipedia.org/wiki/Kernel_(statistics)#Nonparametric_statistics) a $h>0$ je vyhlazovací parametr (kernel bandwidth).

Velmi často se jako jádro vybírá hustota normálního rozdělení ("gaussovka"), vyhlazovací parametr pak odpovídá směrodatné odchylce. 

Vlastnosti jádrového odhadu, volba jádra a další detaily jsou daleko za intencemi předmětu. Spokojme se s tím, že software udělá většinu práce za nás. Demonstrovat si ale můžeme vliv parametru `bw`  v `geom_density()` níže. Zkuste:
- měnit šířku jádra (od 0.05 do 5)
- měnit počet vzorků

```r
bw <- .05
nsamples <- 15
set.seed(60)
x <- rnorm(nsamples)
df <- data.frame(x)
ggplot(df, aes(x=x)) + 
    #geom_histogram(aes(y=..density..), bins=15, colour = 1, fill = "white") +
    geom_density(lwd=2, bw=bw, colour="blue", fill="lightblue", alpha=0.3) +    # Experiment with 'bw' parameter
    stat_function(fun=dnorm, colour="red", lwd=2) +
    geom_rug(x=x, colour="black", lwd=2)
```

#### Q-Q plot: porovnání empirických kvantilů s kvantily vybrané distribuce

Poměrně často potřebujeme ověřit, jak moc se empirická distribuce výběru podobá nějaké vybrané teoretické distribuci, zpravidla normální. Má-li být naše porovnání založeno na vizuální analýze, potom můžeme využít porovnání hodnot empirické a teoretické distribuční funkce či KDE a teoretické hustoty. Velmi populární jsou ale i [Q-Q grafy](https://en.wikipedia.org/wiki/Q%E2%80%93Q_plot), srovnávající kvantiltedyy teoretické distribuce s kvantily výběrovými.  Pokud by data přesně odpovídala vybrané teoretické distribuci, potom by si odpovídaly i kvantily a data tak ležela v grafu na přímce. Z odchylek od přímky lze při jistých zkušenostech vyčíst odlišnosti v šikmosti, špičatosti, přítomnost odlehlých hodnot (outlierů) apod.

```r
library(cowplot)
dfm = 2
x <- rt(150, df=dfm)
df <- data.frame(x)

options(repr.plot.width = 12, repr.plot.height =5)
qq <- ggplot(df, aes(sample=x)) +
            stat_qq(distribution=qnorm) +
            stat_qq_line(distribution=qnorm, col='red')
pdfpdf <- ggplot(data = data.frame(x = c(-5, 5)), aes(x)) +
            stat_function(fun=dt, args=list(df=dfm), size=1, aes(colour="t")) +
            stat_function(fun=dnorm, size=1, aes(colour="N")) +
            geom_histogram(data=df, aes(y=..density..), alpha=0.3) +
            scale_color_manual(name = "", values=c("black", "red"), labels=c("t", "N"))

plot_grid(qq, pdfpdf)
```

### Čeho si všímáme a jak problémům čelit

- **Závislost či nezávislost naměřených hodnot** - v rámci jedné proměnné můžeme mít data závislá (typicky v časových řadách) nebo nezávislá. Zásadní je, pokud máme předpoklad iid, aby byl (co nejvíce) splněn, tj. podmínky při získávání dat se s časem neměnily. K ověření tohoto předpokladu nezávislosti lze použít testy autokorelace, např. [Durbinův-Watsonův test](https://en.wikipedia.org/wiki/Durbin%E2%80%93Watson_statistic) či [Breuschův-Godfreyúv test](https://en.wikipedia.org/wiki/Breusch%E2%80%93Godfrey_test). Případná nežádoucí autokorelace (závislost) dat v rámci jedné veličiny může pro další analýzu představovat velmi zásadní překážku: proces, který data vygeneroval, se totiž měnil v čase a případný statistický model by se tedy správně měl rovněž měnit.
- **Homogenita výběru** - opět souvisí s předpokladem iid, zde s částí *identically distributed*. Typickým problémem jsou odlehlé hodnoty (*outliery*). Ty můžeme z dat buď vyloučit, je-li to oprávněné z podstaty jejich vzniku (např. jde o výpadky měření), nebo je dodatečně modelovat. V každém případ, event. použít metody robustní statistiky (medián namísto průměru, IQR místo variance apod.).
- **Homoskedasticita** - jsou-li data [homoskedastická](https://en.wikipedia.org/wiki/Homoscedasticity_and_heteroscedasticity), pak nevykazují proměnnou varianci. Opakem je [heteroskedasticita](https://en.wikipedia.org/wiki/Homoscedasticity_and_heteroscedasticity), kdy se variance mění např. v závislosti na poloze či čase. Taková situace představuje problémy zejm. v regresní analýze.
- **Normalita výběru** - ačkoliv existuje celá řada populárních rozdělení, značná část statistického aparátu je založena na předpokladu normality. Kromě vizuální analýzy (Q-Q plot, distribuční funkce apod.) můžeme [normalitu testovat](https://en.wikipedia.org/wiki/Normality_test). Populární jsou [Shapirův-Wilkův test](https://en.wikipedia.org/wiki/Shapiro%E2%80%93Wilk_test), [KS test](https://en.wikipedia.org/wiki/Kolmogorov%E2%80%93Smirnov_test) či [Jarque-Berův test](https://en.wikipedia.org/wiki/Jarque%E2%80%93Bera_test).  [Některé distribuce](https://en.wikipedia.org/wiki/Normal_distribution#Related_distributions) lze za určitých podmínek považovat za přibližně normální. Alternativně, dává-li to u konkrétních dat aspoň trochu smysl, můžeme se o normalitu pokusit např. transformacemi. Populární jsou odmocnina, logaritmus a převrácená hodnota. Opatrnost je vždy na místě!
- **Rozsah výběru** - čím více, tím lépe. Jsou-li data koncentrována, potom jich stačí menší počet, než jsou-li více rozptýlena.

## Dvou- a vícerozměrné veličiny (bi- and multivariate analysis)

Uvažujme nyní, že namísto skalární veličiny máme (pro jednoduchost dvourozměrný) sloupcový náhodný vektor $$
\begin{pmatrix}
X \\ Y
\end{pmatrix}.
$$ 

Ten má opět nějaké rozdělení a podobně jako u náhodné veličiny je i zde možné provádět **náhodný výběr** 

$$
\begin{pmatrix}
X_1 \\ Y_1
\end{pmatrix}
\ldots
\begin{pmatrix}
X_n \\ Y_n
\end{pmatrix}
$$

a pracovat s jeho realizacemi, tj. pozorovanými daty. Typickým příkladem náhodného vektoru bývají měření více veličin vztažené k témuž časovému okamžiku, např. 
- GPS poloha cíle v zeměpisné šířce, délce a nadmořské výšce (trojrozměrný vektor)
- kurzovní hodnoty zahraniční měny (vícerozměrný vektor)
- počet dostupných koloběžek v jednotlivých částech města (vícerozměrný vektor)
- systolický a diastolický krevní tlak (dvourozměrný vektor)

Zavedeme výběrový průměrný vektor 

$$
\begin{pmatrix}
\bar{X} \\ \bar{Y}
\end{pmatrix} 
= \frac{1}{n} \sum_{i=1}^{n}
\begin{pmatrix}
X_i \\ Y_i
\end{pmatrix}
$$


### Výběrová kovarianční matice

> [Výběrová kovarianční matice](https://en.wikipedia.org/wiki/Sample_mean_and_covariance) je $n\times n$ pozitivně definitní symetrická matice definovaná
>
>$$
\begin{aligned}
S &= \frac{1}{n-1} \sum_{i=1}^{n}
\left[
\begin{pmatrix}
X_i \\ Y_i
\end{pmatrix}
-
\begin{pmatrix}
\bar{X} \\ \bar{Y}
\end{pmatrix}
\right]
\left[
\begin{pmatrix}
X_i \\ Y_i
\end{pmatrix}
-
\begin{pmatrix}
\bar{X} \\ \bar{Y}
\end{pmatrix}
\right]^\intercal \\
&=
\begin{pmatrix}
S_{X}^2 & S_{XY} \\ S_{XY} & S_{Y}^2
\end{pmatrix},
\end{aligned}
$$
>
>kde $S_X^2$ a $S_Y^2$ jsou výběrové rozptyly $X$ a $Y$. Veličina $S_{XY}$ je skalární výběrová kovariance $X$ a $Y$.

Kovarianční matici používáme v celé řadě úloh, např. v [analýze hlavních komponent (PCA)](https://en.wikipedia.org/wiki/Principal_component_analysis), redukci dimenzionality aj. Kovarianční matice spolu se středním vektorem parametrizují [vícerozměrnou normální distribuci](https://en.wikipedia.org/wiki/Multivariate_normal_distribution). Nám dobře známá jednorozměrná "gaussovka" je tedy jen speciálním případem, kdy náhodný vektor redukujeme na jednorozměrnou veličinu.

### Výběrová korelační matice

>Výběrová [korelační matice](https://en.wikipedia.org/wiki/Correlation#Correlation_matrices) $n\times n$ pozitivně definitní symetrická matice $R$, jejímiž prvky jsou korelační koeficienty
>$$
r_{XY} = \frac{S_{XY}}{S_X S_Y}
$$
>
>a má tedy podobu
>
>$$
R = 
\begin{pmatrix}
r_{XX} & r_{XY} \\ r_{XY} & r_{YY}
\end{pmatrix}
=
\begin{pmatrix}
1 & r_{XY} \\ r_{XY} & 1
\end{pmatrix}.
$$

Pro vyšší rozměry náhodného vektoru platí analogické vztahy. Korelační matici může pěkně ilustrovat korelogram:

```r
.libPaths("~/tmp/bi-prs/")      # Modify lib to whatever necessary
install.packages("corrplot") 
library(corrplot)
R <- cor(mtcars)                # See contextual help on mtcars in JupyterLab
corrplot(R, method="ellipse")    # methods: “circle”, “square”, “ellipse”, “number”, “shade”, “color”, “pie”.
```

### Box plot a houslový (violin) plot

Box plot (též krabicový graf) asi není potřeba představovat. Podstatné je ale uvědomit si, co zobrazuje a jaké má záludnosti. Najdeme v něm *obvykle*:
- medián (nebo průměr!) - ve středu boxu
- horní a dolní kvartil - vymezují box. Ten tedy obsahuje 50% dat a jeho délka je právě IQR.
- whiskery ("vousy") - *obvykle* o délce 1.5 IQR - pozor, není pravidlem!!!
- outliery - co je mimo whiskery

Pokud nevykreslíme data (jitter), potom má boxplot určité záludnosti, typicky v případě dat pocházejících ze směsí, viz příklad. Zároveň to je důvod, proč používat violin plot, který namísto krabičky vykreslí zrcadlené KDE odhady.

```r
data <- data.frame(
    name=c(rep("300x N(0, 20)", 300), rep("150x N(3, 1) & 150x N(-3, 1)", 300), rep("300x N(0, 1)", 500)),
    value=c(rnorm(300, 0, sqrt(20)), rnorm(150, 3, 1), rnorm(150, -3, 1), rnorm(500, 0, 1))
)

options(repr.plot.width = 15, repr.plot.height=5)
ggplot(data, aes(x=name, y=value, fill=name)) +
    geom_boxplot(outlier.color="red", show.legend=F) +
    coord_flip() #+
    #geom_jitter(color="black", size=0.4, alpha=0.9)    # uncomment for jitter
```

```r
options(repr.plot.width=15, repr.plot.height=5)
ggplot(data, aes(x=name, y=value, fill=name)) +
    geom_violin(show.legend=F) +
    coord_flip()
```

### Čeho si všímáme a jak problémům čelit

Každou proměnnou studujeme z hlediska jejích vlastností podobně, jako je uvedeno u jednorozměrných veličin výše. Navíc se ale zajímáme o *míru korelace* mezi jednotlivými veličinami. Pro další analýzy bychom se rádi vyhnuli zejména [**multikolinearitě**](https://en.wikipedia.org/wiki/Multicollinearity), tj. situaci, kdy mezi dvěma veličinami je silná lineární závislost. Typicky v lineární regresi dokáže takový jev nadělat pořádné problémy.

**Příklad multikolinearity**
- automobily - rychlost skutečná a přístrojová (nepřesnost průměru pneumatiky a úmyslně zavedená chyba)
- letectví - výška AGL (above ground level), MSL (mean sea level), FL (flight level)
 
V souvislosti s daty, zejména těmi vícerozměrnými, si všímáme také [chybějících dat](https://en.wikipedia.org/wiki/Missing_data), která by mohla naše analýzy komplikovat. Pokud chybějící data představují překážku, lze je buď [doplňovat (imputation)](https://en.wikipedia.org/wiki/Imputation_(statistics)) nebo vymazávat celá měření. Oba přístupy mají svá pro a proti.
